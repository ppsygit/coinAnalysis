```python
import pandas as pd
df = pd.read_excel('DATA/_기간별평균환율조회(월별)_20240726.xlsx')
df_currency = df[['일자','기준환율']].copy()
df_currency = df_currency.drop(0)
```

```python
df_currency
```

```python
df = pd.read_csv('DATA/upbit_2023_5min.csv')
df
```

```python
df['일자'] = pd.to_datetime(df['candle_date_time_utc'])
df['일자'] = df['일자'].dt.strftime(r'%Y.%m.%d')
```

```python
df_merged = df.merge(df_currency, on='일자', how='left')
df_merged
```

```python
df_merged
```

```python
df_merged.isna().sum()
```

```python
df_filled = df_merged.ffill().fillna('1,267.30').copy()
df_filled['기준환율']= df_filled['기준환율'].str.replace(',','').astype(float)
df_filled['trade_price'] = df_filled['trade_price'].astype(float)
df_filled
```

```python
df_upbit=df_filled.copy()
df_upbit['tradeprice(usdt)'] = df_upbit['trade_price'] / df_upbit['기준환율']
df_upbit
```

```python
df_upbit.to_csv('upbit_2023.csv')
df
```
