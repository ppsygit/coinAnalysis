### 볼린저 밴드 이용 매매

```python
import pandas as pd
```

```python
df = pd.read_csv('DATA/upbit_BTC_prices_2023_days.csv')
```

```python
df['candle_date_time_utc'] = pd.to_datetime(df['candle_date_time_utc'])
df
```

```python
import ta.volatility as tav
df['upper'] = tav.bollinger_hband(df['trade_price'])
df['lower'] = tav.bollinger_lband(df['trade_price'])
df['middle'] = tav.bollinger_mavg(df['trade_price'])
df['width'] = tav.bollinger_wband(df['trade_price'])
```

```python
df[['upper', 'middle', 'lower']]
```

```python
df = df.loc[27:].reset_index().drop(columns=['Unnamed: 0','index'])
df
```

### 볼린저 밴드의 폭이 좁아지는 구간 찾기

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,4))
plt.plot(df['candle_date_time_utc'],df['width'])
```

```python
df['width_slope'] = (df['width'] - df['width'].shift().fillna(df['width'])) /2
df['width_slope'] 
```

```python
df_minus_width_slope = df.loc[df['width_slope']<-2]
len(df_minus_width_slope)
```

```python
df_breakout = df.loc[df['trade_price']>=df['upper']]
len(df_breakout)
```

```python
plt.figure(figsize=(16,15))

ax1 = plt.subplot(3,1,1)
plt.plot(df['candle_date_time_utc'], df['width_slope'], color='black', label='Width_slope')
plt.axhline(y=0, color='black')
plt.axhline(y=-2, color='red')
plt.plot(df_minus_width_slope['candle_date_time_utc'],df_minus_width_slope['width_slope'], 'ro')
# Fill the area where width_slope is positive
plt.fill_between(df['candle_date_time_utc'], df['width_slope'], where=(df['width_slope'] >= 0), color='blue', alpha=0.5, interpolate=True)
# Fill the area where width_slope is negative
plt.fill_between(df['candle_date_time_utc'], df['width_slope'], where=(df['width_slope'] < 0), color='red', alpha=0.5, interpolate=True)
plt.xlabel('Candle Date Time (UTC)')
plt.ylabel('Width slope')
plt.title('Width slope Over Time')
plt.legend()

ax2 = plt.subplot(3,1,2, sharex=ax1)
plt.plot(df['candle_date_time_utc'], df['width'], label='width')
plt.plot(df_minus_width_slope['candle_date_time_utc'],df_minus_width_slope['width'], 'ro')
plt.plot(df_breakout['candle_date_time_utc'], df_breakout['width'], 'go')
plt.xlabel('Candle Date Time (UTC)')
plt.ylabel('Width')
plt.title('Width Over Time')
plt.legend()

ax3 = plt.subplot(3,1,3, sharex=ax1)
plt.plot(df['candle_date_time_utc'], df['trade_price'], color = 'black', label = 'trade_price')
plt.plot(df['candle_date_time_utc'], df['upper'], color = 'r')
plt.plot(df['candle_date_time_utc'], df['lower'], color = 'b')
plt.plot(df_minus_width_slope['candle_date_time_utc'], df_minus_width_slope['trade_price'], 'ro')
plt.plot(df_breakout['candle_date_time_utc'], df_breakout['trade_price'], 'go')
plt.xlabel('Candle Date Time (UTC)')
plt.ylabel('trade_price')
plt.title('trade_price Over Time')
plt.legend()

plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.35)

plt.show()
```
