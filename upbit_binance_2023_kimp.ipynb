### 김프 분석

```python
import pandas as pd
df = pd.read_csv('DATA/2023_BTC-KRW_upbit-binance.csv',index_col=0)
df
```

```python
df_PM = df.copy()
df_PM.loc[df_PM['diff'] < 0, 'sgn'] = '-'
df_PM.loc[df_PM['diff'] >= 0, 'sgn'] = '+'
df_PM
```

```python
df_PM['diff_percentage'].describe()
```

##### 업비트, 바이낸스 매수, 매도 

- 역프 꼈을 때 업비트에서 매수, 바이낸스에서 매도
- 김프 꼈을 때 업비트에서 매도, 바이낸스에서 매수 

```python
upbit_trade_info_all = list()
sgn_stat = 'buy'

for i in range(len(df_PM)):
    if sgn_stat =='buy' and df_PM.loc[i]['diff_percentage']<-0.5:
        trade_info = dict()
        buy_price = df_PM.loc[i]['Upbit']
        sgn_stat='sell'
        trade_info['buy'] = buy_price

    elif sgn_stat =='sell' and df_PM.loc[i]['diff_percentage']>0.5:
        sell_price = df_PM.loc[i]['Upbit']
        sgn_stat = 'buy'
        trade_info['sell'] = sell_price
        upbit_trade_info_all.append(trade_info)
    else:
        continue
upbit_trade_info_all
```

```python
df_upbit_trade_info = pd.DataFrame(upbit_trade_info_all)
df_upbit_trade_info['profit'] = df_upbit_trade_info['sell'] - df_upbit_trade_info['buy']
df_upbit_trade_info
```

```python
df_upbit_trade_info['profit_rate'] = df_upbit_trade_info['profit']/df_upbit_trade_info['sell']*100
df_upbit_trade_info['profit_rate'].mean()
```

```python
# sgn + 일 때 upbit 매도, binance 매수
# sgn - 일 때 binance 매도, upbit 매수

trade_info_all = list()
sgn_stat = 'buy'

for i in range(len(df_PM)):
    if sgn_stat =='buy' and df_PM.loc[i]['diff_percentage']<0:
        trade_info = dict()
        buy_price = df_PM.loc[i]['Binance']
        sgn_stat='sell'
        trade_info['buy'] = buy_price
    
    elif sgn_stat =='sell' and df_PM.loc[i]['diff_percentage']>=0:
        sell_price = df_PM.loc[i]['Binance']
        sgn_stat = 'buy'
        trade_info['sell'] = sell_price
        trade_info_all.append(trade_info)
    else:
        continue
trade_info_all

binance_trade_info = pd.DataFrame(trade_info_all)
binance_trade_info['profit'] = binance_trade_info['sell'] - binance_trade_info['buy']

binance_trade_info['profit_rate'] = binance_trade_info['profit']/binance_trade_info['sell']*100
binance_trade_info['profit_rate'].mean()
```
