```python
import pandas as pd
df_binance = pd.read_csv('DATA/binance_2023_usdt_BTC.csv')
df_binance
```

```python
df_upbit = pd.read_csv('DATA/upbit_BTC_prices_2023.csv')
df_upbit
```

### 바이낸스 USDT를 KRW로 변환

```python
df_binance_krw = df_binance[['Open time','Close']].copy()
df_binance_krw
```

```python
#환율 불러오기
df_currency = pd.read_excel('DATA/_기간별평균환율조회(월별)_20240726.xlsx')
df_currency = df_currency.rename(columns={'일자':'date'})
df_currency
```

```python
df_binance_krw['date'] = pd.to_datetime(df_binance_krw['Open time']).dt.strftime('%Y.%m.%d')
df_binance_krw = df_binance_krw.merge(df_currency[['date', '기준환율' ]], on='date', how='left')
df_binance_krw
```

```python
df_binance_krw['기준환율'] =df_binance_krw['기준환율'].ffill().bfill()
df_binance_krw['Close'] = df_binance_krw['Close'].astype('float')
df_binance_krw['기준환율'] =df_binance_krw['기준환율'].str.replace(',','').astype('float')
df_binance_krw['Close'] = df_binance_krw['Close'] * df_binance_krw['기준환율']
df_binance_krw
```

```python
df_binance_krw = df_binance_krw[['Open time','Close']]
df_binance_krw
```

```python
df_upbit_krw = df_upbit[['candle_date_time_utc','trade_price']]
df_upbit_krw = df_upbit_krw.rename(columns={'candle_date_time_utc' : 'Open time',
                             'trade_price'          : 'Close' })
df_upbit_krw['Open time'] = df_upbit_krw['Open time'].str.replace('T', ' ')
df_upbit_krw['Open time']= pd.to_datetime(df_upbit_krw['Open time'])
df_binance_krw['Open time'] = pd.to_datetime(df_binance_krw['Open time'])
df_krw = df_upbit_krw.merge(df_binance_krw,on='Open time',how='left')
df_krw = df_krw.rename(columns={'Close_x':'Upbit', 'Close_y':'Binance'})
df_krw
```

```python
df_krw['Binance'].isna().sum()
```

```python
df_krw[['Upbit','Binance']].dropna().astype('int')
df_krw['diff'] = df_krw['Upbit'] - df_krw['Binance']
df_krw
```

```python
df_krw['diff_percentage'] = (df_krw['Upbit'] / df_krw['Binance'] - 1) * 100
df_krw
```

### 업비트와 바이낸스 BTC 가격 시각화

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(16, 10))

ub_won_cleaned_copy = df_krw.set_index('Open time')
ub_won_cleaned_copy['Upbit'].plot(label='Upbit')
ub_won_cleaned_copy['Binance'].plot(label='Binance')
ub_won_cleaned_copy['diff'].plot(label='diff')
plt.legend()
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.xlabel("")
plt.title('<2023년도 업비트와 바이낸스에서 비트코인 가격 흐름>')

plt.show()
```

### 업비트와 바이낸스 BTC 시세차이 시각화

```python
df_krw['diff_percentage'].mean()
```

```python
plt.figure(figsize=(16, 10))
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
ax = df_krw['diff_percentage'].plot(label='diff_percentage(%)', color='green')

plt.ylabel("percentage(%)")
plt.title("<2023년도 거래소간 비트코인 시세차이 (업비트-바이낸스)>")
ax.axhline(y=0, color='r', linestyle='--', linewidth=1.5)  # 빨간색 점선, 두께 1.5
ax.text(x=-5000, y=0.1, s='zero', fontsize=15, color='r')
ax.axhline(y=df_krw['diff_percentage'].mean(), color='b', linestyle='--', linewidth=1.5)  # 빨간색 점선, 두께 1.5
ax.text(x=-5000, y=1.8, s='mean', fontsize=15, color='b')

plt.legend(loc='best', ncol=3)
plt.xlabel("")
plt.show()
```
