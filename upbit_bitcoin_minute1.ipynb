```python
import requests
import pandas as pd

def get_1min_stick_of_1day(market,date):
    df2 = pd.DataFrame()
    headers = {"accept": "application/json"}
    
    for i in range(7):
        hour = str((i+1)*3)
        url = f"https://api.upbit.com/v1/candles/minutes/1?market=KRW-{market}&to={date}T{hour.zfill(2)}%3A00%3A00&count=180"
        response = requests.get(url, headers=headers)
        print(response, hour)
        df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')
        df2 = pd.concat([df2,df]).reset_index().drop(columns='index')
    
    hour = str((i+1)*3)
    url = f"https://api.upbit.com/v1/candles/minutes/1?market=KRW-{market}&to={date}T00%3A00%3A00&count=180"
    response = requests.get(url, headers=headers)
    print(response, 24)
    df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')
    df2 = pd.concat([df2,df]).reset_index().drop(columns='index')
   
    return df2
```

```python
from datetime import datetime, timedelta

def date_range(start, end):
    start = datetime.strptime(start, "%Y-%m-%d")
    end = datetime.strptime(end, "%Y-%m-%d")
    dates = [(start + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((end-start).days+1)]
    return dates

dates = date_range("2024-06-01", "2024-07-01")
dates
```
