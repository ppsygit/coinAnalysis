### 하루 5분봉 가져오는 함수

```python
import requests
import pandas as pd

def get_5min_stick_of_1day(coin, date):

    url = f"https://api.upbit.com/v1/candles/minutes/5?market=KRW-{coin}&to={date}T12%3A00%3A00&count=144"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    print(response)
    df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')

    url = f"https://api.upbit.com/v1/candles/minutes/5?market=KRW-{coin}&to={date}T23%3A59%3A00&count=144"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    df2 = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')

    result = pd.concat([df,df2]).reset_index().drop(columns='index')
    return result
```

```python
df = get_5min_stick_of_1day('BTC','2023-06-02')
df
```

### 날짜 리스트 생성 함수

```python
from datetime import datetime, timedelta

def date_range(start, end):
    start = datetime.strptime(start, "%Y-%m-%d")
    end = datetime.strptime(end, "%Y-%m-%d")
    dates = [(start + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((end-start).days+1)]
    return dates

dates = date_range("2024-06-01", "2024-07-01")
dates
```

### 날짜구간 5분봉 가져오는 함수 

```python

def get_5min_stick_of_interval(coin, dates):
    df = pd.DataFrame()
    errorCnt = 0
    for date in dates:
        try:
            df2 = get_5min_stick_of_1day(coin, date)
            df = pd.concat([df,df2])
            print(coin +" "+ date+ '입력')
        except:
            print(coin+" "+ date + '입력실패')
            errorCnt += 1
            if errorCnt == 288:
                break
            else:
                continue
    return df
    # return df.reset_index().drop(columns='index')
```

### 코인 입력, 저장 경로 입력 하여 2023년 정보 csv 파일로 받기

```python
import pandas as pd
# 코인 리스트 가져오기
import requests

url = "https://api.upbit.com/v1/market/all?isDetails=true"

headers = {"accept": "application/json"}

res = requests.get(url, headers=headers)

df = pd.DataFrame(res.json())
df = df.loc[df['market'].str.contains('KRW')]
df['market'] = df['market'].str.strip('KRW-')
coinlist = list(df['market'])
```

```python
def get_2023_to_csv(coin, path):
    dates = date_range("2023-01-01", "2023-12-31")
    df = get_5min_stick_of_interval(coin,dates)
    df.to_csv(path+"upbit_"+coin+"_prices_2023.csv")
```

```python
def get_2022_to_csv(coin, path):
    dates = date_range("2022-01-01", "2022-12-31")
    df = get_5min_stick_of_interval(coin,dates)
    df.to_csv(path+"upbit_"+coin+"_prices_2023.csv")
```
